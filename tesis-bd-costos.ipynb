```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. CARGAR ARCHIVO EXCEL EN DATAFRAME

```python
df =pd.read_excel('tesis-bd-costos.xlsx', sheet_name='BD')
df.columns
df['Máquina'].unique()
```

"""

# 2. FILTROS SOBRE LA BASE DE DATOS 
Se crean funciones que filtran la base de datos según el tipo de máquinas presentes en los artículos científicos de interés.  
El propósito de los filtros es la automatización de la edición de los gráficos de modo que se muestren solo las máquinas correspondientesd al estudio de interés.

```python
# Reducir el df por criterio de máquinas que usa el escenario B de Akay: Volteo con Harvester (incluye procesado) + Forwarder
df_akay = df[(df['Máquina'] == 'Harvester')
                   | (df['Máquina'] == 'Fellerbuncher')
                  | (df['Máquina'] == 'Skidder-Garra')
                | (df['Máquina'] == 'Skidder')
                | (df['Máquina'] == 'Grúa')]
       

df_akay['Máquina'].unique()
```

```python
df = df_akay

plt.figure(figsize=(20,18))

plt.subplot(2,2,1)
sns.stripplot(x='CT, $/hrpl', y= 'Máquina', hue='Fuente', s=10 , alpha=.5, data=df)
plt.title('COSTO HORARIO TOTAL ($/hrpl) por MÁQUINA')

plt.subplot(2,2,2)
sns.stripplot(x='CT, $/m3', y= 'Máquina', s=10 , alpha=.5, hue='Fuente',  data=df)
plt.title('COSTO TOTAL por PRODUCCIÓN ($/m3) por MÁQUINA')

plt.subplot(2,2,3)
sns.stripplot(x='CF, $/hrpl', y= 'Máquina', s=10 , hue='Fuente', alpha=.5 , data=df)
plt.title('COSTOS FIJOS ($/hrpl) por MÁQUINA')
plt.xticks(rotation=45)

plt.subplot(2,2,4)
sns.stripplot(x='CV, $/hrpl', y= 'Máquina', s=10 , hue='Fuente', alpha=.5 , data=df)
plt.title('COSTOS VARIABLES ($/hrpl) por MÁQUINA')


plt.tight_layout()
plt.show()
```
